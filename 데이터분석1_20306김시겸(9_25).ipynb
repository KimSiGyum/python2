- 최고기온의 날짜(max_date) 구하기

```python
for row in data:
    if row[2] == '':
        row[2] = -999
    row[2] = float(row[2])
    print(row)
```

```python
for row in data:
    print(type(row[2]))
```

```python
max_date = ''
max_temp = -999
```

```python
for row in data:
    if row[2] > max_temp:
        max_temp = row[2]
        max_date = row[0]
```

```python
print('기상 관측 아래 서울의 최고 기온이 가장 높았던 날은', max_date, '로', max_temp , '도 였다.')
```

```python
min_temp = 999
min_date = ''

for row in data:
    if row[3] == '':
        row[3] = 999
    row[3] = float(row[3])

    if row[3] < min_temp:
        min_temp = row[3]
        min_date = row[0]

print('기상 관측 아래 서울의 최저 기온이 가장 낮았던 날은', min_date, '로', min_temp, '도 였다.')
```
