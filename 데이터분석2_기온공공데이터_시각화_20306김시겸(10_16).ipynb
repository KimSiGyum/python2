```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,16,32]
z = [3,9,27,81,243]
#제목 추가
plt.title('거듭제곱의 그래프')
#축이름
plt.xlabel('x축')
plt.ylabel('y축')
#꺾은선 그래프, 범례
plt.plot(x,y, label = '2의 거듭제곱')
plt.plot(x,z, label = '3의 거듭제곱')
#범례나타내기
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2 * n for n in x]

#plt.plot(x,y, color = 'r', marker = 'o', linestyle = '-.')
plt.plot(x, y, 'ro-.')

#격자 추가
plt.grid()
#x축 y축 범위 지정
plt.axis([0, 6, 0, 12])
plt.show()
```
