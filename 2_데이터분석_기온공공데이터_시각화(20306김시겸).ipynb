```python
#csv 파일 읽어오기

import csv
f = open('weather.csv', encoding = 'euc-kr')
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
print(header)
type(data)
```

```python
days = []
for i in range(31):
    days.append([])

for row in data:
    if row[2] != '' and row[0].split('-')[1] == '08':
        days[int(row[0].split('-')[2]) - 1].append(float(row[2]))

plt.style.use('ggplot')
plt.figure(figsize=(10,4))
plt.boxplot(days, showfliers=False)

plt.show()
```

```python
month=[[0] for n in range(12)]

#for i in range(12):
#    month.append([])

for row in data:
    if row[2] != '' and row[3] != '':
        temp = float(row[2]) - float(row[3])
        month[int(row[0].split('-')[1]) - 1].append(temp)
print(month)
plt.figure(figsize = (10,4))

plt.boxplot(month, showfliers = False)
plt.show()
```

### 나이의 결측값 처리하기
- 결측값 나이의 평균값으로 대체하기

```python
print(data[5])
```

```python
#편균나이 구하기
sum = 0
count = 0
avg_age = 0
for row in data:
    if row[5] != '':
        sum += float(row[5])
        count += 1
avg_age = sum / count

print('평균나이는', int(avg_age))
```

```python
for row in data:
    if row[5] == '':
        row[5] = int(avg_age)
print(data[5])
```

1) 타이타닉의 사망자, 생존자를 딕셔너리에 저장하여 BarChart로 표현하가

```python
survived = {'0':0, '1':0}
print(header)
for row in data:
    if row[1] == '1':
        survived['1'] += 1
    else:
        survived['0'] += 1
print(survived)
plt.bar(survived.keys(), survived.values())
plt.show()
```

```python
print(header)
```

```python
SBG = {'생존 남성':0, '사망 남성':0, '생존 여성':0, '사망 여성':0}

for row in data:
    if row[4] == 'male':
        if row[1] == '1':
            SBG['생존 남성'] += 1
        else:
            SBG['사망 남성'] += 1
    else:
        if row[1] == '1':
            SBG['생존 여성'] += 1
        else:
            SBG['사망 여성'] += 1

print(SBG)
```

```python
plt.bar(SBG.keys(), SBG.values(), color = ['b', 'orange', 'b', 'orange'])
plt.show()
```

```python
plt.pie(SBG.values(), labels=SBG.keys())
plt.show()
```
