[문제] 8월 15일의 최고기온, 최저기온 데이터 그래프 그리기

```python
high = []

for row in data:
    if row[2] != '':
        if row[0].split('-')[1] == '08':
            high.append(float(row[2]))

```

```python
high = []
low = []
for row in data:
    if row[2] != '' and row[3] != '':
        if row[0].split('-')[1] == '08' and row[0].split('-')[2] == '15':
            high.append(float(row[2]))
            low.append(float(row[3]))
```

```python
plt.figure(figsize = (15, 4))
plt.plot(high, 'hotpink', label = '최고기온')
plt.plot(low, 'skyblue', label = '최저기온')
plt.legend()
plt.show()
```

[문제] 1960~2020년까지 평균기온 그래프 그리기

```python
#plt.plot(년도, 년도별 평균기온)
#years = [1960, 1961, 1962, ......, 2020]
#avg = [1960년 평균기온의 평균, 1961년 평균기온의 평균, ......, 2020년 평균기온의 평균]
```

```python
avg = []
years = [1960]
sum = 0
days = 0

for row in data:
    #년도가 같은지 => 같은해 평균기온의 합, 날짜수 카운팅
    if int(row[0].split('-')[0]) == years[-1]:
        if row[0] != '':
            sum += float(row[1])
            days += 1
    #년도가 다른면 => 평균기온의 평균구하기
    elif int(row[0].split('-')[0]) > years[-1]:
        #print(years[-1], '년도는 ', days, '일이었다')
        avg.append(sum/days)
    #바뀐 년도가 마지막 년도 (2021)이면 끝
        if int(row[0].split('-')[0]) == 2021:
             break
    #아니면 평균기온의 합을 초기화, 날짜수 1로 초기화
        else:
              years.append(years[-1]+1)
              sum = float(row[1])
              days = 1

plt.figure(figsize = (8, 3))
plt.title('1960~2020년의 평균 기온 그래프')
plt.xlabel('연도')
plt.ylabel('기온')
plt.plot(years, avg, 'red')
plt.show()
```

```python
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[2] != '':
        if month == '08':
            aug.append(float(row[2]))
        if month == '01':
            jan.append(float(row[2]))

plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```
